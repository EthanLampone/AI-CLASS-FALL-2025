```python
import time ## USE THIS TO ADD SOME SPIC TO THE FILLING OF THE JUGS
            ## MAKE IT SEEM LIKE IT TAKING TIME TO FILL/REFILL THEM ***
```

```python
## CREATE A CLASS FOR THE WATER JUG PUZZLE
class WaterJugPuzzle():

  def __init__(self):
    ## INITIALIZE TWO EMPTY SETS FOR THE 5 LITER JUG AND 3 LITER JUG, RE
    self.five_lit = set([])
    self.three_lit = set([])

  def both_jugs_are_empty(self): # LOOK TO SEE IF BOTH JUGS ARE EMPTY
    return len(self.five_lit) == 0 and len(self.three_lit) == 0

  def both_jugs_are_full(self):
    return len(self.five_lit) == 5 and len(self.three_lit) == 3

  def three_lit_is_full(self): ## LOOK TO SEE IF THREE LITER JUG IS FULL
    return len(self.three_lit) == 3

  def five_lit_is_full(self): ## LOOK TO SEE IF FIVE LITER JUG IS FULL
      return len(self.five_lit) == 5

  def three_lit_is_empty(self): ## LOOK TO SEE IF THREE LITER JUG IS EMPTY
    return len(self.three_lit) == 0

  def five_lit_is_empty(self): ## LOOK TO SEE IF FIVE LITER JUG IS EMPTY
    return len(self.five_lit) == 0

  def undo_state(self): ## FUNCTION TO UNDO THE STATE OF EITHER FIVE/THREE JUGS
    self.five_lit = set(list(self._five_lit))
    self.three_lit = set(list(self._three_lit))

  def save_state(self): ## FUNCTION TO SAVE THE CURRENT STATE OF FIVE/THREE JUGS
    self._five_lit = set(list(self.five_lit))
    self._three_lit = set(list(self.three_lit))

  def check_state(self): ## CHECK STATES OF BOTH FULL AND/OR BOTH EMPTY
    return not (self.both_jugs_are_empty() or self.both_jugs_are_full())

## --------------------------------------------------------------------
# RULE 1: FILL ONE OF THE JUGS COMPLETELY (5 LITER)
## --------------------------------------------------------------------

  def R1_fill_five_liter_jug(self):

    self.save_state()

    if not self.five_lit: ## EVALUATES TO True IF "five_lit" IS EMPTY

      self.five_lit.add('1')
      self.five_lit.add('2')
      self.five_lit.add('3')
      self.five_lit.add('4')
      self.five_lit.add('5')

    ## CHECKS TO ENSURE BOTH JUGS ARE NOT EMPTY OR BOTH FULL -
    ## CANT FILL ONE JUG OR THE OTHER IF BOTH ARE FULL/EMPTY!
    if not self.check_state(): self.undo_state()

## --------------------------------------------------------------------
# RULE 2: FILL THE OTHER JUG WITH WATER UNTILL IT IS FULL
# THIS SHOULD RESULT IN THE FIVE LITER JUG HAVING 2 LITERS LEFT &
# THE THREE LITER JUG BEING COMPLETELY FULL
## --------------------------------------------------------------------

  def R2_fill_three_lit_jug_from_five_lit_jug(self):

    self.save_state()

    if not self.three_lit and self.five_lit_is_full:
      self.five_lit.remove('5')
      self.five_lit.remove('4')
      self.five_lit.remove('3')
      self.three_lit.add('1')
      self.three_lit.add('2')
      self.three_lit.add('3')

    if not self.check_state(): self.undo_state()

## --------------------------------------------------------------------
# RULE 3: CHECK TO SEE IF THE THIRD JUG IS FULL. IF FULL, EMPTY THE
## JUG COMPLETELY
## --------------------------------------------------------------------

  def R3_empty_three_lit_jug(self):

    self.save_state()

    if self.three_lit_is_full:
      self.three_lit = set([])

    if not self.check_state(): self.undo_state()

## --------------------------------------------------------------------
# RULE 4: WITH THE 3 LITER WATER JUG NOW EMPTY, WE CAN FILL THE REST
# OF THE FIVE LITER JUG INTO THE NOW EMPTY THREE LITER JUG
## --------------------------------------------------------------------

  def R4_fill_rest_of_three_lit_jug(self):

    self.save_state()

    if self.three_lit_is_empty and self.five_lit_is_empty:
      self.five_lit.remove('2')
      self.five_lit.remove('1')
      self.three_lit.add('1')
      self.three_lit.add('2')

    if not self.check_state(): self.undo_state()

## --------------------------------------------------------------------
# RULE 5: WITH THE THREE LITER JUG NOW FILLED WITH 2 LITERS AND THE
# FIVE LITER JUG IS NOW EMPTY, WE REFILL THE FIVE LITER JUG AGAIN
## --------------------------------------------------------------------

  # *** RECALL R1_fill_five_liter_jug() FOR THIS STEP ***

## --------------------------------------------------------------------
# RULE 6: TO END, WE ADD ONE LITER FROM THE NEWLY REFILLED FIVE JUG
# INTO THE THREE LITER JUG
## --------------------------------------------------------------------

  def R6_add_one_liter_to_three_liter_jug(self):

    self.save_state()

    if self.five_lit_is_full and len(self.three_lit) == 2:
      self.three_lit.add('3')
      self.five_lit.remove('5')

    if not self.check_state(): self.undo_state()

## --------------------------------------------------------------------
# EXTRA: DISPLAY HOW FULL/EMPTY THE JUGS ARE
## --------------------------------------------------------------------

  def print_state(self):
    print(f"===> Five-liter jug: {self.five_lit}")
    print(f"===> Three-liter jug: {self.three_lit}")
    time.sleep(1)
```
